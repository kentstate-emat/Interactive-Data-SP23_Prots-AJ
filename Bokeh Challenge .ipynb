```python
import numpy as np
import pandas as pd
from bokeh.models import HoverTool
from bokeh.plotting import figure, show, reset_output
from bokeh.io import output_notebook, output_file
```

```python
twitch = pd.read_csv('https://raw.githubusercontent.com/Hamidest/InteractiveData_EMAT/main/twitchdata-update.csv')
twitch.head()
```

```python
twitch_new = twitch.groupby('Language')
twitch_new.count()
```

```python
counts = twitch_new.count()
counts = counts.sort_values('Channel', ascending = False)
counts 
```

```python
x = list(counts.index)
y = counts['Channel']
output_notebook()
output_file('in-class-figure.html')

p1 = figure (x_range = x, sizing_mode = 'stretch_width')
p1.line(x, y)

show(p1)
```
